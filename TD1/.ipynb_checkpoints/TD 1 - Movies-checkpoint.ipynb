# [Movies dataset](https://perso.telecom-paristech.fr/eagan/class/igr204/datasets)

```python
import numpy as np
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data = pd.read_csv("datasets/film.csv",sep=";", encoding='ISO-8859-1')
data.head()
```

```python
data.dtypes
```

```python
data.shape
```

Etude de la répartition des Awdars en fonction du temps des films

```python
data.groupby(data.Awards).Length.describe()
```

Quel est l'évolution du nombre de film sortie par année ?

```python
plt.rcParams['figure.figsize'] = [15, 6]
plt.rcParams['figure.dpi'] = 200
data.groupby(data.Year).size().plot()
```

Quel est le type avec le plus grand nombre de film ?

```python
data.groupby(data.Subject).size().sort_values()[::-1]
```

Quel est le type de film qui a le plus d'award ?

```python
data[(data.Awards == 'Yes')].groupby(data.Subject).size().sort_values()[-1:]
```

Quel est le TOP 5 des films avec le plus de popularité ?

```python
data[(data.Popularity.isnull()==False)].sort_values(by="Popularity")[::-1][:5]
```

Quel directeur a fait le plus de film ?

```python
data.groupby(data.Director).size().sort_values()[::-1]
```

Quel sont les films d'Alfred Hitchcock du plus vieux au plus récent ?

```python
data[(data.Director == 'Hitchcock, Alfred')].sort_values(by="Year")
```

Quel est le nombre de film sortie par an en fonction de son type ?

```python
data.pivot_table('Title',index='Year',columns='Subject',aggfunc=pd.Series.nunique)
```

Quel est l'évolution du temps moyen de film au cours du temps ?

```python
data.groupby("Year").Length.mean().plot()
```
